### Backup original results (hdf5 and log files) before fine-tuning

#### Make folders

```python
Folder1 = './Results'
Folder2 = './Logs'
Folder3 = './Config'

Folder4 = './Benchmarks/Results'
Folder5 = './Benchmarks/Logs'
Folder6 = './Benchmarks/Config'

Sig1 = '/ART'
Sig2 = '/II'

FolderName = '/240216'

# Create the folder
os.makedirs(Folder1 + Sig1 + FolderName, exist_ok=True)
os.makedirs(Folder1 + Sig2 + FolderName, exist_ok=True)
os.makedirs(Folder2 + Sig1 + FolderName, exist_ok=True)
os.makedirs(Folder2 + Sig2 + FolderName, exist_ok=True)
os.makedirs(Folder3 +FolderName, exist_ok=True)

os.makedirs(Folder4 + Sig1 + FolderName, exist_ok=True)
os.makedirs(Folder4 + Sig2 + FolderName, exist_ok=True)
os.makedirs(Folder5 + Sig1 + FolderName, exist_ok=True)
os.makedirs(Folder5 + Sig2 + FolderName, exist_ok=True)
os.makedirs(Folder6 + FolderName, exist_ok=True)
```

#### Copy and paste the original files into the new folders

```python
def BackUp(Folder, FolderName, Sig=None):
    # Construct source and target directories
    if Sig is not None:
        source_directory = Folder + Sig
        target_directory = Folder + Sig + FolderName
    else:
        source_directory = Folder
        target_directory = Folder + FolderName
    
    # Ensure the target directory exists
    os.makedirs(target_directory, exist_ok=True)
    
    for item in os.listdir(source_directory):
        source_path = os.path.join(source_directory, item)
        target_path = os.path.join(target_directory, item)
        
        # If the item is a file (not a directory) and does not exist in the target, copy it
        if os.path.isfile(source_path) and not os.path.exists(target_path):
            shutil.copy(source_path, target_path)
            print(f"Copied: {item}")
        elif os.path.isfile(source_path):
            print(f"Skipped (already exists): {item}")

BackUp(Folder1, FolderName, Sig=Sig1)
BackUp(Folder1, FolderName, Sig=Sig2)
BackUp(Folder2, FolderName, Sig=Sig1)
BackUp(Folder2, FolderName, Sig=Sig2)
BackUp(Folder3, FolderName, Sig=None)


BackUp(Folder4, FolderName, Sig=Sig1)
BackUp(Folder4, FolderName, Sig=Sig2)
BackUp(Folder5, FolderName, Sig=Sig1)
BackUp(Folder5, FolderName, Sig=Sig2)
BackUp(Folder6, FolderName, Sig=None)
```

### Restore original results (hdf5 and log files) before fine-tuning

```python
Folder1 = './Results'
Folder2 = './Logs'
Folder3 = './Config'

Folder4 = './Benchmarks/Results'
Folder5 = './Benchmarks/Logs'
Folder6 = './Benchmarks/Config'

Sig1 = '/ART'
Sig2 = '/II'

FolderName = '/Original'
```

```python
import shutil
import os

def Restore(Folder, FolderName, Sig=None):

    # Construct source and target directories
    if Sig is not None:
        source_directory = Folder + Sig + FolderName
        target_directory = Folder + Sig 
    else:
        source_directory = Folder + FolderName
        target_directory = Folder 

    # Get the list of files from the source folder
    for filename in os.listdir(source_directory):

        if '.ipynb_checkpoints' in filename:
            continue;
            
        # Create the full path for source and target files
        source_file = os.path.join(source_directory, filename)
        target_file = os.path.join(target_directory, filename)
        
        # Copy the file (with metadata)
        shutil.copy2(source_file, target_directory)
        print(f'Copied "{filename}" to {target_directory}')
    
    print('All files have been copied successfully.')

# Example usage
Restore(Folder1, FolderName, Sig=Sig1)
Restore(Folder1, FolderName, Sig=Sig2)
Restore(Folder2, FolderName, Sig=Sig1)
Restore(Folder2, FolderName, Sig=Sig2)
Restore(Folder3, FolderName, Sig=None)


Restore(Folder4, FolderName, Sig=Sig1)
Restore(Folder4, FolderName, Sig=Sig2)
Restore(Folder5, FolderName, Sig=Sig1)
Restore(Folder5, FolderName, Sig=Sig2)
Restore(Folder6, FolderName, Sig=None)
```
