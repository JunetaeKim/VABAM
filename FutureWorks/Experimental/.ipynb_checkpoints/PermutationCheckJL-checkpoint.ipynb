```python
import numpy as np
```

```python
def KernelGen (SortedWindow, MinIdx, WindowSize):
    
    MinIdx -=1
    
    TotalSpace = WindowSize-1
    LeftSpaceSize = MinIdx - 0 
    RightSpaceSize = WindowSize-LeftSpaceSize-1
    InterIDX = 1


    LeftSide = []
    RightSide = []

    # Kernel generation
    while LeftSpaceSize>0 or RightSpaceSize>0:

        if LeftSpaceSize>0:
            LeftSide.append(SortedWindow[InterIDX])
            InterIDX += 1
            LeftSpaceSize -= 1

        if RightSpaceSize>0:
            RightSide.append(SortedWindow[InterIDX])
            InterIDX += 1
            RightSpaceSize -= 1

    Kernel = np.concatenate([LeftSide[::-1], [SortedWindow[0]], RightSide])    
    
    return Kernel
```

```python
def ZigZagPerm (K,N):

    SortedWindow = []

    SortedWindow.append([K, 1])
    for i in range(1, N+1):

        if i < min(K, N-K+1):

            SortedWindow.append([K-i, 2*i])
            SortedWindow.append([K+i, 2*i+1])

        else:

            SortedWindow.append([K-i, min(K, N-K+1)+i])
            SortedWindow.append([K+i, min(K, N-K+1)+i])

    SortedWindow = np.array(SortedWindow)        
    ## Conditional filter
    SortedWindow = SortedWindow[(SortedWindow[:, 0] >0) & (SortedWindow[:, 0] <=N)].copy()
    SortedWindow = SortedWindow[np.argsort(SortedWindow[:, 0])][:, 1].copy()
    return SortedWindow
```

```python
ResCheckSum = 0

for N in range(2, 30):
    for K in range(1, N):
        
        Window = np.arange(N)+ 1
        ResKernel = KernelGen(Window, K, N)
        TestKernel = ZigZagPerm(K, N)
        
        print('True:', ResKernel.astype(np.int32))
        print('Test:', TestKernel.astype(np.int32))
        print()

        ResCheck = np.sum(ResKernel - TestKernel)
        ResCheckSum += ResCheck
        
```
