```python
import numpy as np
```

```python
def KernelGen (SortedWindow, MinIdx, WindowSize):
    
    MinIdx -=1
    
    TotalSpace = WindowSize-1
    LeftSpaceSize = MinIdx - 0 
    RightSpaceSize = WindowSize-LeftSpaceSize-1
    InterIDX = 1


    LeftSide = []
    RightSide = []

    # Kernel generation
    while LeftSpaceSize>0 or RightSpaceSize>0:

        if LeftSpaceSize>0:
            LeftSide.append(SortedWindow[InterIDX])
            InterIDX += 1
            LeftSpaceSize -= 1

        if RightSpaceSize>0:
            RightSide.append(SortedWindow[InterIDX])
            InterIDX += 1
            RightSpaceSize -= 1

    Kernel = np.concatenate([LeftSide[::-1], [SortedWindow[0]], RightSide])    
    
    return Kernel
```

```python
def Test (N, K):
    
    SortedWindow = np.arange(N)+ 1
    ResKernel = KernelGen(SortedWindow, K, N)
    
    Ln = K -1 
    Rn = N - K
    Ln, Rn


    ### Left side
    if Ln <= Rn:
        S = min(Ln, Rn)
    elif Ln > Rn:
        S = min(Ln, Rn) + 1

    J1 = np.arange(S+1)
    LeftA = 2*J1

    J2 = np.arange(Ln-S+1)
    LeftB = 2*S + J2

    ULeft = np.union1d(LeftA, LeftB) # 합집합
    Left = ULeft[ULeft >0] # 0 제거
    Left = np.sort(Left)[::-1] # Descending sort


    ### Right side
    S = min(Ln, Rn)

    J1 = np.arange(S+1)
    RightA = 2*J1+1

    J2 = np.arange(Rn-S+1)
    RightB = 2*S + J2 + 1

    URight = np.union1d(RightA, RightB) # 합집합
    Right = URight[URight >0] # 0 제거
    Right = np.sort(Right) # Ascending sort

    TestKernel = np.concatenate([Left, Right])
    
    print('True:', ResKernel.astype(np.int32))
    print('Test:', TestKernel.astype(np.int32))
    print()
    
    Check = np.sum(ResKernel - TestKernel)
    return Check


```

```python
N=2
K=2
Test (N, K) 
```

```python
ResCheckSum = 0

for N in range(2, 30):
    for K in range(1, N):
        #print('N:', N, 'K:', K, 'Check:', Test (N, K) )
        #print(Test (N, K) )
        
        ResCheck = Test(N, K) 
        print('N:', N, 'K:', K, 'Check:', ResCheck )
        ResCheckSum += ResCheck
        
```
